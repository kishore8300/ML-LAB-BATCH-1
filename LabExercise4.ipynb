```python
import numpy as np
A=np.array([[3,2,2],[2,3,-2]])

print("Original Array:",A)
u,s,vt=np.linalg.svd(A)
print("U:\n",u)
print("\nS:\n",s)
print("VT:\n",vt)
print("\nDiagonal :",np.diag(A))
```
