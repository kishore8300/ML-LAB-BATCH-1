```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

data = {
    'Fever': ['Yes', 'No', 'No', 'Yes', 'No', 'No'],
    'Nausea': ['Yes', 'Yes', 'No', 'No', 'Yes', 'No'],
    'Missed': ['Yes', 'Yes', 'Yes', 'No', 'No', 'No'],
    'infection': ['Yes', 'Yes', 'Yes', 'No', 'No', 'No']
}

df = pd.DataFrame(data)

# Encode categorical yes/no as 1/0
for col in ['Fever', 'Nausea', 'Missed', 'infection']:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

X = df[['Fever', 'Nausea', 'Missed']]
y = df['infection']

# Train decision tree
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)

# Plot tree
plt.figure(figsize=(8,6))
plot_tree(clf, feature_names=X.columns, class_names=['Not infected', 'infected'], filled=True)
plt.title(" Decision Tree using ID3")
plt.show()

```

```python

import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

data = pd.read_csv("diabetes.csv")

df = pd.DataFrame(data)

feature_cols = ['Glucose', 'Insulin', 'BloodPressure']
X = df[feature_cols]
y = df['Outcome']

clf = DecisionTreeClassifier(criterion='entropy', max_depth=2)
clf.fit(X, y)

class_names = [str(c) for c in sorted(y.unique())]

plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=feature_cols, class_names=class_names, rounded=True)
plt.title("C4.5-like Decision Tree")
plt.show()
```
