```python
#1.3
from numpy.linalg import det, eig

nums = np.array([[2, 4], [1, 3]])

print("Matrix:\n", nums)
print("Determinant:", det(nums))

eigenvalues, eigenvectors = eig(nums)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```
