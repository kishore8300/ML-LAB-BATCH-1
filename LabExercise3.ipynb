```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import linear_model

saledata ={
    'week':[1,2,3,4,5],
    'sales':[1.2,1.8,2.6,3.2,3.8]
}
df=pd.DataFrame(saledata,columns=['week','sales'])
print(df)

plt.scatter(df["week"], df["sales"], color="green")
plt.title("Regression among week and sales")
plt.xlabel("X - axis - Week")
plt.ylabel("Y- Dependent - Sales")

week=df['week'].values.reshape(1,-1)
sales=df['sales'].values.reshape(1,-1)
print(week)
print(sales)

X = df[["week"]]
y = df["sales"]

regr=linear_model.LinearRegression()
regr.fit(X,y)

print("Intercept \n",regr.intercept_)
print("coefficient \n",regr.coef_)

# plt.plot(week,regr.predict(week),color="red",linewidth=4)


print("\nThe Regression Equation is",regr.coef_,"* X+",regr.intercept_)

# Fit the model for the given data
weeks_to_predict = pd.DataFrame({'week': [7, 9]})
predicted_sales = regr.predict(weeks_to_predict)
print("\nPredicted sales for week 7:", predicted_sales[0])
print("Predicted sales for week 9:", predicted_sales[1])
pred = regr.predict(X)
plt.plot(X,pred,marker="*")

# Compute Adjusted R squared Error
print("\nAdjusted R Squared for Regression model:",regr.score(X,y))
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import linear_model

saledata ={
    'week':[1,2,3,4,5],
    'sales':[12,18,22,28,35]
}
df=pd.DataFrame(saledata,columns=['week','sales'])
print(df)

plt.scatter(df["week"], df["sales"], color="green")
plt.title("Regression among week (hours) and Studying")
plt.xlabel("X - axis - Week")
plt.ylabel("Y- Dependent - Sales")

week=df['week'].values.reshape(-1,1)
sales=df['sales'].values.reshape(-1,1)
print(week)
print(sales)

X = df[["week"]]
y = df["sales"]

regr=linear_model.LinearRegression()
regr.fit(X,y)

print("Intercept \n",regr.intercept_)
print("coefficient \n",regr.coef_)



print("\nThe Regression Equation is",regr.coef_,"* X+",regr.intercept_)


pred = regr.predict(X)
plt.plot(X,pred,color="red",marker="*")

#predicting sales for week 7 and week 9
weeks_to_predict = pd.DataFrame({'week': [7, 9]})
predicted_sales = regr.predict(weeks_to_predict)
print("\nPredicted sales for week 7:", predicted_sales[0])
print("Predicted sales for week 9:", predicted_sales[1])


print("\nAdjusted R Squared for Regression model:",regr.score(X,y))
```
